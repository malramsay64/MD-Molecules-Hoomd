```python
import hoomd
from hoomd import md
import molecule
import basic
import quaternion
import numpy as np
from altair import Chart, X, Bin
import pandas
```

```python
initial_file = './Trimer-13.50-1.30.gsd'
steps = 100
snap1, fin1, snap2, fin2 = basic.main(initial_file, steps)
```

```python
nbodies = snap1.particles.body.max() + 1
angmom = np.abs(quaternion.as_quat_array(snap1.particles.angmom[:nbodies].astype(float)))
mom = np.sqrt(np.power(snap1.particles.velocity[:nbodies], 2).sum(axis=1))*3
momentum_stats = pandas.DataFrame(data={'angmom':angmom, 'momentum':mom})
```

```python
Chart(momentum_stats).mark_bar().encode(
    x=X('angmom',
        bin=Bin(
            maxbins=50.0,
        ),
    ),
    y='count(*):Q',
)
```

```python
Chart(momentum_stats).mark_bar().encode(
    x=X('momentum',
        bin=Bin(
            maxbins=50.0,
        ),
    ),
    y='count(*):Q',
)
```
